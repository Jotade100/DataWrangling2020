```python
import numpy as np
import pandas as pd
heroes = pd.read_csv("heroes_information.csv")
```

```python
###instalar desde anaconda navigator######
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

### 1
Se puede saber con la vista

```python
query2 = " SELECT * FROM heroes limit 10"
```

```python
pysqldf(query2)
```

### 2

```python
query2 = " SELECT COUNT(DISTINCT(publisher)) FROM heroes"
print(query2)
pysqldf(query2)
```

### 3

```python
query2 = " SELECT COUNT(*) FROM heroes WHERE Height > 200"
print(query2)
pysqldf(query2)
```

### 4

```python
query2 = " SELECT * FROM heroes WHERE Height >= 200 AND race LIKE 'Human'"
print(query2)
pysqldf(query2)
```

### 5

```python
query2 = " SELECT * FROM heroes WHERE Weight < 200 and Weight > 100"
print(query2)
pysqldf(query2)
```

### 6

```python
query2 = " SELECT COUNT(*) FROM heroes WHERE `Eye color` LIKE 'blue' OR `Eye color` LIKE 'red'"
print(query2)
pysqldf(query2)
```

### 7

```python
query2 = " SELECT * FROM heroes WHERE (Race in ('Human', 'Mutant') AND `Hair color` LIKE 'green') OR (Race = 'Vampire' AND `Hair color` like 'black')"
print(query2)
pysqldf(query2)
```

### 8

```python
query2 = " SELECT * FROM heroes ORDER BY Race DESC"
print(query2)
pysqldf(query2)
```

### 9

```python
query2 = " SELECT Gender, COUNT(*) FROM heroes GROUP BY `Gender` "
print(query2)
pysqldf(query2)
```

### 10

```python
query2 = " SELECT Publisher, COUNT(*) FROM heroes GROUP BY `Publisher` HAVING COUNT(*) > 15 "
print(query2)
pysqldf(query2)
```
